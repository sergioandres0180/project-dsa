# Objetivo
Realizar la limpieza de los datos, realizando transformaciones necesarias antes de entrenar el modelo.


```python
# Librerias a emplear
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Primeras filas del DataFrame, información general (tipos de datos, número de filas y columnas)
# estadísticas descriptivas de las columnas numéricas.

df_anuncios = pd.read_csv('data_clean_anuncios.csv')

display(df_anuncios.head())
display(df_anuncios.info())
display(df_anuncios.describe())
```

```python
## Identificacion y gestion de datos faltantes
missing_values = df_anuncios.isnull().sum()
print("Número de valores faltantes por columna:")
print(missing_values)

missing_percentage = (missing_values / len(df_anuncios)) * 100
print("\nPorcentaje de valores faltantes por columna:")
print(missing_percentage)
```

## Propuestas para el manejo de valores faltantes por columna:
- 'bano': Todos los valores faltantes. Se propone eliminar esta columna.
- 'city': 0.61% valores faltantes. Se propone eliminar esta columna, dado que la columna 'Ciudad' proporciona información ya normalizada sin faltantes.
- 'Estado': 1.52% valores faltantes. Se propone eliminar esta columna, ya que la normalizacion de estos datos no ofrecerán info relevante para su uso.

```python
#cuando "city" es nulo
df_anuncios[df_anuncios['city'].isna()]

#eliminar filas cuando city es nulo
df_anuncios = df_anuncios.dropna(subset=['city'])
```

```python
#escogiendo columnas utiles para el modelo
columnas_utiles = ['valor_cop', 'area', 'Cuartos', 'tipo', 'city', 'Ciudad']
df_anuncios = df_anuncios[columnas_utiles]
```

### Limpieza de columna 'tipo'

```python
# Pasar todo a minúsculas y eliminar espacios
df_anuncios['tipo'] = df_anuncios['tipo'].str.strip().str.lower()

df_anuncios['tipo'] = df_anuncios['tipo'].replace({
    'apartamento ': 'apartamento',
    'apto': 'apartamento',
    'apartamentoo': 'apartamento',
    'csa': 'casa'
})

#filtrar solo por apartamentos y casas, ya que para el resto de inmuebles no se tiene suficiente información
df_anuncios = df_anuncios[df_anuncios['tipo'].isin(['apartamento', 'casa'])]

print(df_anuncios['tipo'].value_counts())
```

### Limpieza columna 'city'

```python
#Convertir a string por si hay valores numéricos o nulos
df_anuncios['city'] = df_anuncios['city'].astype(str)

#Pasar todo a minúsculas y eliminar espacios al inicio y final
df_anuncios['city'] = df_anuncios['city'].str.strip().str.lower()

#Eliminar saltos de línea, tabs y espacios extra
df_anuncios['city'] = df_anuncios['city'].str.replace(r'\s+', ' ', regex=True)

#Extraer solo la primera palabra antes de "Unidades" o del salto de línea original
df_anuncios['city'] = df_anuncios['city'].str.extract(r'^([\w\sáéíóúÁÉÍÓÚñÑ]+)', expand=False)  

# Eliminar espacios al inicio y final
df_anuncios['city'] = df_anuncios['city'].str.strip()

# Pasar a formato título (Antioquia → Antioquia)
df_anuncios['city'] = df_anuncios['city'].str.title()

# Extraer la parte inicial que parece un departamento o ciudad
df_anuncios['city'] = df_anuncios['city'].str.extract(
    r'^([A-Za-zÁÉÍÓÚáéíóúÑñ\s]+?)(?:\s+Unidades|\s+\d|\n|$)',
    expand=False
)

#Contar la frecuencia de cada valor
frecuencias = df_anuncios['city'].value_counts()

# Filtrar solo los valores que aparecen al menos 5 veces
valores_comunes = frecuencias[frecuencias >= 5].index

df_anuncios = df_anuncios[df_anuncios['city'].isin(valores_comunes)]

#Revisar los valores que quedaron
pd.set_option('display.max_rows', None)
print(df_anuncios['city'].value_counts())

```

### Limpieza de la variable ciudad

```python
#Agrupando barrios poco frecuentes, como minimo tener 5 datos
umbral = 5

# Contar frecuencia de cada barrio dentro de cada ciudad
frecuencias_barrio_ciudad = df_anuncios.groupby('city')['Ciudad'].value_counts()

# Función para limpiar barrios raros
def limpiar_barrio(row):
    try:
        if frecuencias_barrio_ciudad[row['city']][row['Ciudad']] >= umbral:
            return row['Ciudad']
        else:
            return 'Otros'
    except KeyError:
        return 'Otros'  # por si hay algún valor raro
       
df_anuncios['barrio_limpio'] = df_anuncios.apply(limpiar_barrio, axis=1)

#Realizando target encoding
# Calcular promedio de precio por ciudad y barrio_limpio
target_mean = df_anuncios.groupby(['city', 'barrio_limpio'])['valor_cop'].mean()

# Mapear el encoding al DataFrame
df_anuncios['barrio_target_enc'] = df_anuncios.apply(
    lambda x: target_mean[x['city'], x['barrio_limpio']], axis=1
)

pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)
print(df_anuncios[['city', 'barrio_limpio']].value_counts().head(200))
```

### Limpieza variables area y cuartos


```python
# Revisión de valores nulos
print(df_anuncios[['area', 'Cuartos']].isnull().sum())

print("Valores únicos de Cuartos:", df_anuncios['Cuartos'].unique())
print("Valores únicos de area:", df_anuncios['area'].unique()[:20])  
```

# Resumen

- Para la columna **city** fue necesario eliminar los valores nulos, que correspondían al 0.6% de los datos. Se verificó que estas filas no contenían información relevante de ubicación geográfica adicional, por lo que se decidió eliminar todas las filas con valores nulos en esta columna. Posteriormente, se realizó una limpieza con expresiones regulares, eliminando espacios innecesarios y dejando solo la primera palabra o expresiones relevantes antes de un salto de línea, debido a que esta variable contenía información extra del anuncio. Se eliminaron filas donde no fue posible realizar la limpieza correctamente.

- Para la columna **tipo** se observó que los tipos "apartamento" y "casa" representaban el 95% de los registros, mientras que el resto de tipos de inmueble contaba con muy poca información. Por lo tanto, se decidió eliminar los registros correspondientes a los tipos minoritarios y trabajar únicamente con estas dos categorías principales.

- Para la columna **Ciudad**, que corresponde principalmente a la zona o barrio del inmueble, se detectó una alta variabilidad. Se procedió a agrupar los barrios por ciudad, creando una nueva columna llamada barrio_limpio, donde si un barrio tenía menos de 5 registros dentro de su ciudad se asignaba la categoría "Otros", y si tenía 5 o más registros se mantenía su valor original.

- Se creó una nueva columna **barrio_target_enc**, mediante Target Encoding, calculando el promedio de valor_cop por combinación de ciudad y barrio limpio. Esta columna permite reflejar de manera numérica la información de la ubicación del inmueble, facilitando su uso en modelos de predicción de precios.

- Para las columnas **area** y **Cuartos** no fue necesaria ninguna transformación, ya que los valores se encontraban consistentes y listos para su uso como variables numéricas en los modelos.
